## How does K-means work? 

Conditions points must satisfy: 
1. each point must belong to a cluster
2. each point can only belong to one cluster (no single point can belong to multiple clusters). 

Steps: 

1. start with unlabeled data 
2. choose number of clusters to create (k-value) (randomly choose k-distinct data point)
3. assign each remaining point to the nearest "cluster" point by calculating the distance to the nearest cluster.   
4. calculate the center of the cluster points (means value of each point vector). <br>
4b. recalculate new cluster center. 
5. assign each point to the nearest cluster center.<br>
5b. assign points to nearest cluster center. 
6. repeat step 4 and 5 until there is no more cluster assignments or until the previous iteration equals the current iteration 

some **consideration** we need to make:
1. how do we choose a reasonable value for k number of clusters?
2. can we evaluate how good our current k value is at determing clusters?

## K-means Clustering Model

```python
# because we're dealing with unsupervised learning we don't have to worry about 
# multicollinarity 

X = pd.get_dummies(df) # assigns numerical value for categorical data columns 
```

```python
X.head()
```

```python
# remember that we have to scale data because we're dealing with distance
# so we want to scale to prevent scaling 

from sklearn.preprocessing import StandardScaler 
```

```python
scalar = StandardScaler()
```

```python
scaled_X = scalar.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
Kmeans_model = KMeans(n_clusters = 2)  
```

```python
# fit computes kmeans cluster vs. fit_predict that computes cluster centers
# we want fit_predict() fitting and returning the correct cluster labels 
# similar to fit_transform
# can break the step to model.fit and model.predict but because
# we're dealing with unsupervised learning we can do both on one step since we're not forecasting
cluster_labels = Kmeans_model.fit_predict(scaled_X)
```

```python
# will swap labels if we rerun fit_predict
# remember labels of cluster are arbitrary.
# the important thing is the same rows are being clustered to the same clusters 
cluster_labels  
```

```python
# we can then join the labels back to original labels 
X['Cluster'] = cluster_labels
```

```python
X # last column has cluster labels 
```
