# Digital Image Analysis 

When you read in an image you read (720, 1280, 3), which can be 720 height, 1280 width, and 3 color intensity in it's standardized RGB values 0 to 255. 
A grascaled image is 0 to 255 or if it's normalized 0 to 1. 

But can we use clustering for images? In any given image there are a gradient of colors, if we wanted to display 6 distinct colors for maybe a computer with limited resources or compress the image.

Within every image there are three channels red, green, and blue. We can reshape the image to an X array feature set where we reformat the 3D array to a 2D array. We can convert a 3D array to a 2D array by mapping H x W to a column  and RGB to each row. The reason we want a 2D array is because clustering works in the 2D space.

Each cluster center is the avg of RGB values sampled from the image. K, the cluster centers, is represented as the number of colors we would like to minimize to, this is called color quantizations 

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg # image class that will read in images 
```

```python
palm_tree_image = mpimg.imread('../data/palm_trees.jpg') # returns a 3D array of the image 
```

```python
# plt.figure(dpi = 200) can strech image 
plt.imshow(palm_tree_image) # imshow needs numpy array can't display directly from filepath
```

```python
# (Height, Width, Color channel) to 2D (H*W,C)
```

```python
(h,w,c) = palm_tree_image.shape
```

```python
c
```

```python
palm_tree_image_2d = palm_tree_image.reshape(h*w,c)
palm_tree_image_2d.shape
```

```python
len(palm_tree_image.shape) # how to check if array is 3d
```

```python
len(palm_tree_image_2d.shape) # how to check if array is 2d
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 16) # I want to quantize image to 6 avg colors 
```

```python
cluster_labels = model.fit_predict(palm_tree_image_2d) # returns an array 
# array will function as an index to reassign the original 2d array to color codes. 
```

```python
cluster_labels # the numbers themselves are meaningless what matters if the pts are matching to same clusters
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int) # these values are the 6th averaged 
```

```python
rgb_codes # it's a 16 by 3 array because there are 16 clusters and each have 3 rgb itensity values 
rgb_codes.shape
```

```python
quantized_image = np.reshape(rgb_codes[cluster_labels], (h,w,c)) 
# filter original dataframe to grab color and then reshape to 3d 
# quantized image as an array 
```

```python
plt.imshow(quantized_image)
```
